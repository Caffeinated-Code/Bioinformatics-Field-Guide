# Bash Tutorial: Inspect A Tiny Bioinformatics Project

**End goal:** build a small QC snapshot from a sample sheet and a FASTA file.

You will practice paths, `head`, `wc`, `grep`, `cut`, `sort`, `uniq`, and `awk`. This notebook uses the Python kernel, but Bash commands run inside cells that start with `%%bash`.

## 1. Find the data

A bioinformatics project usually begins with files. First, confirm where you are and list the tutorial data.

```python
%%bash
pwd
ls -lh ../data
```

## 2. Inspect the sample sheet

Use `head` before opening any unfamiliar file. Your first job is to learn the columns without changing the file.

```python
%%bash
head ../data/samples.tsv
echo
wc -l ../data/samples.tsv
```

## 3. Count samples by condition

This is a classic first-pass metadata check: do the groups look balanced?

```python
%%bash
tail -n +2 ../data/samples.tsv | cut -f2 | sort | uniq -c
```

## 4. Inspect FASTA records

FASTA headers start with `>`. Count the records and print the gene labels.

```python
%%bash
grep -c '^>' ../data/mini_transcripts.fasta
grep '^>' ../data/mini_transcripts.fasta
```

## 5. Calculate sequence lengths

This `awk` command joins sequence lines until the next header, then reports each sequence length.

```python
%%bash
awk '/^>/{if(seq){print name "\t" length(seq)}; name=substr($0,2); seq=""; next} {seq=seq $0} END{print name "\t" length(seq)}' ../data/mini_transcripts.fasta
```
